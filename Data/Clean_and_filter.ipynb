```python
#Identify all the columns that are not similar amongst the different leagues
def similarHeadersAmongLeagues():
    for ll in all_leagues:
        for y_l in ll:
            epl_l = list(y_l.columns)
        
            for l in all_leagues:
                temp_bin = []
                #print(y.Date[0])
                y = l[0]
                for col in epl_l:
                    if col not in y.columns:
                        #latest.remove(col)
                        temp_bin.append(col)
                print(temp_bin)
    return 

similarHeadersAmongLeagues()
```

```python
#Drop all columns that is not included in the above
def dropUnnamedCols(cols):
    for l in all_leagues:
        for y in l:
            for col in y.columns:
                if col not in cols:
                    y.drop(col,axis = 1,inplace=True)
        
dropUnnamedCols(cols_to_keep)
```

```python
#Do a simple check that all the columns of all the frames match
def checkIfAllColumnsMatch():
    allMatch = False
    max_len =0
    max_set = []
    for l_1 in all_leagues:
        for y_1 in l_1:
            test = list(y_1.columns)
            for l in all_leagues:
                for y in l:
                    y_c = list(y.columns)
                    if len(set(y_c).difference(set(test))) >0:
                        if len(set(y_c).difference(set(test)))>max_len:
                            max_len = len(set(y_c).difference(set(test)))
                            max_set = set(y_c).difference(set(test))
                        print("NOT MATCHING")
                    else:
                        allMatch = True
    print(max_set)
    return allMatch
checkIfAllColumnsMatch()
```

```python
all_leagues = [b_ligas,epls,laligas,ligue1s,serieas]
league_names = ['Bundelsliga','EPL','LaLiga','Ligue1','SerieA']
all_years = [b_y,e_y,la_y,li_y,s_y]
```

```python
#Save all the dataframes as well as the totals
df_all = pd.DataFrame()
for l,n,y in zip(all_leagues,league_names,all_years):
    df = pd.DataFrame()
    league_name = str(n)
    for y_name,l_y in zip(y,l):
        year_name = str(y_name)
        path = os.getcwd()+'/Similar_years_leagues/'+league_name+'/'+year_name+'.pickle'
        l_y.to_pickle(path)
        l_y['season'] = year_name
        df = pd.concat([df, l_y], ignore_index=True)
    path = os.getcwd()+'/Similar_years_leagues/'+league_name+'/All.pickle'
    df.to_pickle(path)
    df['League'] = league_name
    df_all = pd.concat([df_all,df],ignore_index = True)
path = os.getcwd()+'/Similar_years_leagues/All.pickle'
df_all.to_pickle(path)
```
